```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
pk = pd.read_csv('pokemon.csv')
```

Посмотрим, что от чего зависит

```python
plt.figure(figsize=(12,12))
sns.heatmap(pk.corr())
plt.show()
```

сильная отрицательная корреляция exp_yeild and catch_rate
средненькая отрицательная корреляция веса и счастья
сильная положительная корреляция между весом и ростом

Построим линейную регрессию catch_rate(exp_yield)

```python
sns.regplot(pk.exp_yield, pk.catch_rate)
plt.show()
```

```python
pk.sort_values('exp_yield', ascending=False)[['name', 'exp_yield', 'catch_rate']].head()
```

Покемон, за которого дают больше всего экспы, имеет высокий catch_rate.
